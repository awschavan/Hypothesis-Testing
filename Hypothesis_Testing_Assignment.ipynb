## Q.1
 A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.


```python
import numpy as np
import pandas as pd
df = pd.read_csv("Cutlets.csv")
df
```

```python
#    Two sample test
#    H0 = There is no difference in the average dia of the cutlet between two units.
#    H1 = There is a difference in the average dia of the cutlet between two units.
```

```python
from scipy import stats
Zcal, P_value = stats.ttest_ind(df["Unit A"], df["Unit B"],equal_var=True)
Zcal, P_value
```

```python
print("Z statistic value:",Zcal)
print("P Value",P_value)
```

```python
if P_value < 0.05:
    print("H0 is rejected and H1 is accepted")
else:
    print("H1 is rejected and H0 accepted")
```

# Q.2
A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.

# Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.

```python
import pandas as pd
import numpy as np
df1 = pd.read_csv("LabTAT.csv")
df1
```

```python
''' Make it in single list with group '''
df1_single_data = pd.melt(df1, value_vars=["Laboratory 1","Laboratory 2","Laboratory 3","Laboratory 4"], var_name='Lab', value_name='TAT')
df1_single_data
```

```python
#    Annova test
#    H0 = there is no difference in the average Turn Around Time (TAT) of reports of the laboratories
#    H1 = there is any difference in the average Turn Around Time (TAT) of reports of the laboratories
```

```python
from statsmodels.formula.api import ols
anova1 = ols('TAT ~ C(Lab)',data=df1_single_data).fit()
anova1
```

```python
import statsmodels.api as sm
table = sm.stats.anova_lm(anova1, type=1)
table
```

```python
pvalue =table['PR(>F)'][0]
pvalue.round(9)
alpha = 0.05
```

```python
if pvalue < alpha:
    print("H0 is rejected and H1 is accepted")
else:
    print("H1 is rejected and H0 is accepted")
```
